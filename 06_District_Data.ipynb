```python
import requests
import pandas as pd
```

```python
r = requests.get('https://api.covid19india.org/state_district_wise.json')
resp = r.json()
```

```python
states = resp.keys()
districts = []
for state in states:
    ddata = resp[state]["districtData"]
    for dist_name in ddata:
        districts.append({
            "name": dist_name,
            "active": ddata[dist_name]["active"],
            "confirmed": ddata[dist_name]["confirmed"],
            "deceased": ddata[dist_name]["deceased"],
            "recovered": ddata[dist_name]["recovered"],
        })

df = pd.DataFrame(districts)
```

```python
df.sample()
```

```python
cnf_sum = df['confirmed'].sum()
cnf_sum
```

```python
pct_df = df[['name', 'confirmed']][:]
pct_df['confirmed_pct'] = 100 * pct_df['confirmed'] / cnf_sum
pct_df.sort_values(by='confirmed_pct', ascending=False).head(20)
```
